This piece of program will prepare the whole data set, including predictors and predicted variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

```

```python
df = pd.DataFrame([])

for i in range(0,12):
    dff = pd.read_csv('/home/cccr/supriyo/panini/data_jgrjd/20CRV3/station_ano_1836-2015_'+str(i)+'.csv')
    dff['date'] = pd.to_datetime(dff['Unnamed: 0'])
    dff.drop(['Unnamed: 0','date'],axis=1)
    dff.index = dff.date
    dff = dff.iloc[:,1:2]
    
    df = pd.concat([df,dff],1)
```
