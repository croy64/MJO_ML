```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import keras

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,Activation, Dropout
import seaborn as sns

from keras.wrappers.scikit_learn import KerasRegressor
from keras import backend as K
```

```python
df = pd.DataFrame([])

for i in range(0,12):
    dff = pd.read_csv('/home/cccr/supriyo/panini/data_jgrjd/20CRV2/station_ano_1979-2012_'+str(i)+'.csv')
    dff['date'] = pd.to_datetime(dff['Unnamed: 0'])
    dff.drop(['Unnamed: 0','date'],axis=1)
    dff.index = dff.date
    dff = dff.iloc[:,1:2]
    
    df = pd.concat([df,dff],1)
```

```python
df2 = pd.DataFrame([])

for i in range(0,12):
    dff2 = pd.read_csv('/home/cccr/supriyo/panini/data_jgrjd/20CRV2/station_ano_1946-1978_'+str(i)+'.csv')
    dff2['date'] = pd.to_datetime(dff2['Unnamed: 0'])
    dff2.drop(['Unnamed: 0','date'],axis=1)
    dff2.index = dff2.date
    dff2 = dff2.iloc[:,1:2]
    
    df2 = pd.concat([df2,dff2],1)
```

```python
df3=pd.concat([df2,df],0)
df3.to_csv('full_data_pressure_20CRV2.txt')
```
